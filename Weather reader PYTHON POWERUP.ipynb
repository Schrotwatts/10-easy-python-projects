```python
import pyttsx3
schrotwatts = pyttsx3.init()
# schrotwatts is the NLP talking model
```

```python
schrotwatts.say("Hello All") # runs so fast that the compiler doesnt give time for schrotwatts to say anything
schrotwatts.runAndWait() #give time to speak
```

```python
#customize the voice
voices = schrotwatts.getProperty("voices")
```

```python
voices
```

```python
schrotwatts.setProperty("voices", voices[0].id) # to select the second voice
```

```python
schrotwatts.say("Nice to meet you")
schrotwatts.runAndWait()
```

```python
#customize the speed(increasing here)
schrotwatts.setProperty("rate", 180);
```

```python
schrotwatts.say("Nice to meet you")
schrotwatts.runAndWait()
```

```python
#customize the speed(increasing here)
schrotwatts.setProperty("volume", 1);
schrotwatts.say("Nice to meet you")
schrotwatts.runAndWait()
```
